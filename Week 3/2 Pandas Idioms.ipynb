```python
import pandas as pd

df = pd.read_csv('../Data Files/census.csv')
df
```

```python
# Method Chainning

(df.where(df['SUMLEV'] == 50) # yang 'SUMLEV' nya 50
.dropna() # drop/hilangkan data na
.set_index(['STNAME', 'CTYNAME']) # Buat index
.rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})) # Rename columns
```

```python
# Hapus row yang SUMLEV == 50 (ADA INDEX DI DROP())
# Rename kolom 'CTYNAME' jadi 'city name'

df.drop(df[df['SUMLEV'] == 50].index).rename(columns={'CTYNAME': 'city name'})
```

```python
df = df[df['SUMLEV'] == 50]
df.set_index(['STNAME', 'CTYNAME'], inplace=True)
df.rename(columns={'ESTIMATEBASE2010': 'Estimate Base 2010'})
df
```

```python
# Buat fungsi min dan max POPESTIMATE2010 sampai POPESTIMATE2015

import numpy as np

def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    return pd.Series({'min': np.min(data), 'max': np.max(data)})
    
df.apply(min_max, axis=1)
```

```python
# Fungsi min max dari column terpilih

import numpy as np
def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    row['max'] = np.max(data)
    row['min'] = np.min(data)
    return row
df.apply(min_max, axis=1)
```

```python
# Maksimal dari column terpilih

rows = ['POPESTIMATE2010',
        'POPESTIMATE2011',
        'POPESTIMATE2012',
        'POPESTIMATE2013',
        'POPESTIMATE2014',
        'POPESTIMATE2015']
df.apply(lambda x: np.max(x[rows]), axis=1)
```
