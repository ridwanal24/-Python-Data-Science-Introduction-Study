```python
import pandas as pd
import numpy as np

df = pd.read_csv('../Data Files/cars.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
nissan = df[df['Make'] == 'NISSAN']
plt.plot(nissan['YEAR'].values, nissan['CITY (kWh/100 km)'])
plt.show()
```

```python
# Pivot table

df.pivot_table(values='(kW)', index='YEAR', columns='Make', aggfunc=np.mean)
```
