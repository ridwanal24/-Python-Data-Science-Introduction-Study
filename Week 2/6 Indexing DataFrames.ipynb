```python
import pandas as pd

df = pd.read_csv('../Data Files/olympics.csv', index_col = 0, skiprows=1)

for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold' + col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver' + col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze' + col[4:]}, inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#' + col[1:]}, inplace=True) 

df.head()
```

```python
# Buat kolom bernama 'country' yang isinya dari kolom index (Afgaistan, algeria, etc)

df['country'] = df.index
df
```

```python
# Buat kolom 'Gold' menjadi index yang baru

df = df.set_index('Gold')
df
```

```python
# Reset index menjadi nomor

df = df.reset_index()
df
```
